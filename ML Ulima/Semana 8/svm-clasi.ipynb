```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

import matplotlib.pyplot as plt
import seaborn as sns

iris = datasets.load_iris()
X = iris.data
Y = iris.target
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=42)
```

```python
svm_model = SVC(kernel='linear')
svm_model.fit(X_train, Y_train)
```

```python
y_pred = svm_model.predict(X_test)
print(f"Acc: {accuracy_score(Y_test, y_pred)}")
```

```python
from sklearn.datasets import make_regression
from sklearn.svm import SVR

X, Y = make_regression(n_samples=1000, n_features=1, noise=25)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=42)
```

```python
svr_model = SVR(kernel='rbf', gamma=0.6)
svr_model.fit(X_train, Y_train)
```

```python
from sklearn.metrics import mean_squared_error
y_pred_reg = svr_model.predict(X_test)
print(f"MSE: {mean_squared_error(Y_test, y_pred_reg)}")
```

```python
plt.scatter(X_test, Y_test, color="blue")
plt.scatter(X_test, y_pred_reg, color="red")
```

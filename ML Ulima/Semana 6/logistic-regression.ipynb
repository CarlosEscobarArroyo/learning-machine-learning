```python
import pandas as pd
import numpy as np
```

```python
num_data = 10000
horas_estudio = np.random.uniform(0,10,10000)
practicas_resueltas = np.random.uniform(0,50,10000)

aprobado = (horas_estudio>4) & (practicas_resueltas>24)
aprobado = aprobado.astype(int)
```

```python
data = pd.DataFrame({
    'horas': horas_estudio,
    'problemas': practicas_resueltas,
    'aprueba': aprobado
})
```

```python
features = ["horas", "problemas"]
X = data[features]
y = data["aprueba"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
X_test["Aprueba"] = model.predict(X_test)
```

```python
X_test[X_test["Aprueba"] == 1]
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
accuracy_score(y_test, X_test['Aprueba'])
```
